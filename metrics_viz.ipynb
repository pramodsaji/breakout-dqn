```python
# Import necessary modules
import json
import matplotlib.pyplot as plt
```

```python
# Function to plot the results
# episodes_array: array of episodes
# y_data: array of y data
# title1: title of the plot
# title2: title of the second plot, if available
# xlabel: label of the x axis
# ylabel: label of the y axis
# is_show_second_line: decide whether to show the second line
def plot_results(episodes_array, y_data, title1, title2, xlabel, ylabel, is_show_second_line = False):
    plt.figure(figsize=(16, 6))
    plt.plot(episodes_array, y_data, label=title1)
    if(is_show_second_line):
        plt.plot(episodes_array, label=title2)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title1)
    plt.legend()
    plt.show()
```

```python
# Function to plot the results from files
# files: array of files to be plotted
def plot_results_from_files(files):
    for index, file in enumerate(files):
        with open(file) as f:
            # Load the data
            data = json.load(f)

            # Extract variables
            episodes = data.get('episodes')
            rewards = data.get('rewards')
            scores = data.get('scores')
            averages = data.get('averages')

            # Plot the results
            plot_results(episodes, scores, f"Scores per Episode - Run #{index + 1}", "", "Episode", "Score", is_show_second_line = False)
            plot_results(episodes, rewards, f"Rewards per Episode - Run #{index + 1}", "", "Episode", "Rewards", is_show_second_line = False)
            plot_results(episodes, averages, f"Average per Episode - Run #{index + 1}", "", "Episode", "Average", is_show_second_line = False)
```
